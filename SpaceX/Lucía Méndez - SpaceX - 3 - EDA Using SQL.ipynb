## Tarea 2: Queries con SQL

### Ejercicio 1

##### Muestra las diferentes plataformas de lanzamiento:

```python
#Rellena la variable query con tu consulta
query = "SELECT DISTINCT LaunchSite FROM SPACEXTBL"
df1  = pd.read_sql(query, con = engine )
df1
```

### Ejercicio 2

##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
query = "SELECT * FROM SPACEXTBL WHERE LaunchSite Like 'CCS%' LIMIT 5"
df2  = pd.read_sql(query, con = engine )
df2
```

### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
query = "SELECT  Customers, SUM(PayloadMass) AS Total_PayloadMass FROM SPACEXTBL WHERE Customers LIKE '%NASA (CRS)%' GROUP BY Customers"
df3  = pd.read_sql(query, con = engine )
df3
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
query = "SELECT Serial As Serial_B1, AVG(PayloadMass) AS Mean_PayloadMass FROM SPACEXTBL WHERE Serial LIKE 'B1%'"
df4  = pd.read_sql(query, con = engine)
df4
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
query = "SELECT MIN(Date) AS First_successful_landing FROM SPACEXTBL WHERE LandingOutcome = 'True RTLS'"
df5  = pd.read_sql(query, con = engine)
df5
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
query = "SELECT Serial, LandingOutcome, PayloadMass FROM SPACEXTBL WHERE LandingOutcome = 'True ASDS' AND PayloadMass BETWEEN 4000 AND 6000"
df6  = pd.read_sql(query, con = engine)
df6
```

### Ejercicio 7

##### Muestra el numero de lanzamientos que completaron su mision correctamente y aquellos que fallaron

```python
query = "SELECT DISTINCT Class, COUNT( Class)  FROM SPACEXTBL GROUP BY Class"
df7  = pd.read_sql(query, con = engine)
df7
```

### Ejercicio 8

##### Muestra las series de los cohetes que transportaron la masa máxima

```python
query = "SELECT Serial, PayloadMass FROM SPACEXTBL WHERE PayloadMass = (SELECT MAX(PayloadMass) FROM SPACEXTBL)"

df8  = pd.read_sql(query, con = engine)
df8
```

### Ejercicio 9

##### Muestra los el resultados del aterrizaje, la serie del cohete y las plataforma de lanzamiento de aquellos lanzamientos fallidos o no intentados en el año 2015


```python
query = "SELECT LandingOutcome, Serial, LaunchSite FROM SPACEXTBL WHERE LandingOutcome LIKE 'None%' OR LandingOutcome LIKE 'False%' AND Date LIKE '2015%"
df9  = pd.read_sql(query, con = engine)
df9
```
